这是一个计算年终数值的demo


一、 读取execel中的数据

```python
import numpy as np
import pandas as pd

df_part1 = pd.read_excel(r"part1.xlsx")
df_part2 = pd.read_excel(r"part2.xlsx")
```

1. 查看读取的dataframe（数据表格）

```python
"""直接使用df0可以查看整个表格。使用head()只显示头部5行"""
df_part1.head()
```

```python
df_part2.head()
```

2. 表格合并，组建一个大表格

```python
df = df_part1.append(df_part2)
df
```

```python
"""教师编号排序"""
df1=df.sort_values(by=["教师编号"])
"""以教师编号为索引"""
df2=df1.set_index("教师编号")
```

```python
df2
```

二、 增加年终奖字段

2.1 基本公式

```python
"""因子1"""
df2["因子1"] = 1
df2.loc[df2["职称"] == "副教授", "因子1"] = 3
df2.loc[df2["职称"] == "教授", "因子1"] = 5
df2.loc[df2["职称"] == "院士", "因子1"] = 9
df2
```

```python
"""因子2, 入职的月份数量"""
df2["月份数"] = (2020 - df2["入职年"]) * 12 + (12 - df2["入职月"])
df2.loc[df2["月份数"] > 10, "因子2"] = 3
df2.loc[df2["月份数"] > 100, "因子2"] = 5
df2.loc[df2["月份数"] > 200, "因子2"] = 9
df2
```

```python
"""计算年终奖"""
df2["年终奖"] = df2["因子1"] * 100 + df2["因子2"] * 980 + 500
df2
```

2.2 个别人调整

```python
"""按教师编号查找"""
df2.loc[155]
```

```python
df2.loc[155,"年终奖"] = df2.loc[155,"年终奖"] + 1000
df2.loc[155]
```

```python
"""按人名调整"""
df2.loc[df2["姓名"] == "苏6","年终奖"] = df2.loc[df2["姓名"] == "苏6","年终奖"] + 1000
df2
```

```python
"""特征1 在50---100区间 增加 120% """
df2.loc[(df2["特征1"] > 50) & (df2["特征1"] < 100), "年终奖"] = df2.loc[(df2["特征1"] > 50) & (df2["特征1"] < 100), "年终奖"] * 1.2
df2
```

三、 年终奖总额调整

```python
"""计算总额"""
df2["年终奖"].sum()
```

```python
"""与大盘比较"""
DaPan = 800000
diff = DaPan / df2["年终奖"].sum()
diff
```

```python
"""年终奖缩放"""
df2["年终奖"] = df2["年终奖"] * diff
```

```python
"""重新计算总额"""
df2["年终奖"].sum()
```
